## Read data Minutes

```python
def Get_aggregated_data(game_id, player_id,parameter_code):
    if game_id == '':
        game_id = '%'
    if player_id == '':
        player_id = '%'
    
    
    sql="select ID,DISTANCE,STARTTIME,ENDTIME,GAME_ID,PARAMETER_CODE,PLAYER_ID,CLUB_ID,HALF_INDICATOR\
        , DIFFERENCE_START_END_PRR_IND,ORDER_COLUMN\
          from aggregated_game_data_sub_t t1\
          where player_id not in (select player_id from keepers_games_t)\
          and player_id in ( select distinct player_id \
                             from aggregated_game_data_sub_t t2\
                             where order_column=1\
                             and  half_indicator = 'S'\
                             and t1.game_id = t2.game_id)\
           and difference_start_end_prr_ind like 'T'\
           and parameter_code like :parameter_code\
           and game_id like :game_id\
           and player_id like :player_id\
           order by parameter_code,half_indicator, order_column asc"
         
    df = pd.read_sql(sql,params={'game_id':game_id,'player_id':player_id, 'parameter_code' :parameter_code}, con=engine)
    return df
```

```python
def Get_aggregated_data_rnd(game_id, player_id,parameter_code):
    if game_id == '':
        game_id = '%'
    if player_id == '':
        player_id = '%'
    
    
    sql="select ID,DISTANCE,STARTTIME,ENDTIME,GAME_ID,PARAMETER_CODE,PLAYER_ID,CLUB_ID,HALF_INDICATOR\
        , DIFFERENCE_START_END_PRR_IND,ORDER_COLUMN\
          from aggregated_game_data_sub_t t1\
          where player_id not in (select player_id from keepers_games_t)\
          and difference_start_end_prr_ind like 'T'\
          and parameter_code like :parameter_code\
          and game_id like :game_id\
          and player_id like :player_id\
          order by parameter_code,half_indicator, order_column asc"
         
    df = pd.read_sql(sql,params={'game_id':game_id,'player_id':player_id, 'parameter_code' :parameter_code}, con=engine)
    return df
```

```python
def Get_aggregated_data_rnd_avg(game_id, player_id,parameter_code):
    if game_id == '':
        game_id = '%'
    if player_id == '':
        player_id = '%'
    
    
    sql="select avg(DISTANCE) as Distance_substitute,ORDER_COLUMN\
          from aggregated_game_data_sub_t t1\
          where player_id not in (select player_id from keepers_games_t)\
          and difference_start_end_prr_ind like 'T'\
          and half_indicator like 'S'\
          and parameter_code like :parameter_code\
          and game_id like :game_id\
          and player_id like :player_id\
          group by order_column\
          order by order_column asc"
         
    df = pd.read_sql(sql,params={'game_id':game_id,'player_id':player_id, 'parameter_code' :parameter_code}, con=engine)
    return df
```

```python
#aggregated data entire game

def Get_aggregated_data_eg(game_id, player_id,parameter_code, whole_game, exclude_ball):
    if game_id == '':
        game_id = '%'
    if player_id == '':
        player_id = '%'
    
    
    #Include ball
    if exclude_ball == 'N':
        sql="select ID,DISTANCE,STARTTIME,ENDTIME,GAME_ID,PARAMETER_CODE,PLAYER_ID,CLUB_ID,HALF_INDICATOR\
        , DIFFERENCE_START_END_PRR_IND,ORDER_COLUMN\
        from aggregated_game_data_t \
        where GAME_ID like :game_id \
        and PLAYER_ID like :player_id\
        and player_id<>400\
        and PARAMETER_CODE like : parameter_code\
        and difference_start_end_prr_ind like '%T%' \
        and full_game_ind like :whole_game \
        order by club_id,player_id,starttime"
    else:
        sql="select ID,DISTANCE,STARTTIME,ENDTIME,GAME_ID,PARAMETER_CODE,PLAYER_ID,CLUB_ID,HALF_INDICATOR\
        , DIFFERENCE_START_END_PRR_IND, ORDER_COLUMN\
        from aggregated_game_data_t t1\
        where GAME_ID like :game_id \
        and PLAYER_ID like :player_id \
        and PARAMETER_CODE like : parameter_code\
        and difference_start_end_prr_ind like '%T%' \
        and PLAYER_ID not in (SELECT ID FROM PLAYERS_T t3 WHERE t3.NAME LIKE 'ball' or NAME LIKE 'Referee' or name like 'centroid')\
        and player_id not in (select player_id from keepers_games_t t2 where t1.game_id = t2.game_id )\
        and full_game_ind like :whole_game\
        and game_id<>754\
        and game_id <>640\
        order by club_id,player_id,starttime"
         
    df = pd.read_sql(sql,params={'game_id':game_id,'player_id':player_id, 'parameter_code' :parameter_code,'whole_game' :whole_game}, con=engine)
    return df
```

```python
#aggregated data entire game

def Get_aggregated_data_eg_avg(game_id, player_id,parameter_code, whole_game, exclude_ball):
    if game_id == '':
        game_id = '%'
    if player_id == '':
        player_id = '%'
    
    
    #Include ball
    if exclude_ball == 'N':
        sql="select ID,DISTANCE,STARTTIME,ENDTIME,GAME_ID,PARAMETER_CODE,PLAYER_ID,CLUB_ID,HALF_INDICATOR\
        , DIFFERENCE_START_END_PRR_IND,ORDER_COLUMN\
        from aggregated_game_data_t \
        where GAME_ID like :game_id \
        and PLAYER_ID like :player_id\
        and player_id<>400\
        and PARAMETER_CODE like : parameter_code\
        and difference_start_end_prr_ind like '%T%' \
        and full_game_ind like :whole_game \
        order by club_id,player_id,starttime"
    else:
        sql="select avg(DISTANCE) as Distance_Entire_game,ORDER_COLUMN\
        from aggregated_game_data_t t1\
        where GAME_ID like :game_id \
        and PLAYER_ID like :player_id \
        and PARAMETER_CODE like : parameter_code\
        and difference_start_end_prr_ind like '%T%' \
        and PLAYER_ID not in (SELECT ID FROM PLAYERS_T t3 WHERE t3.NAME LIKE 'ball' or NAME LIKE 'Referee' or name like 'centroid')\
        and player_id not in (select player_id from keepers_games_t t2 where t1.game_id = t2.game_id )\
        and full_game_ind like :whole_game\
        and game_id<>754\
        and game_id <>640\
        group by order_column\
        order by order_column"
         
    df = pd.read_sql(sql,params={'game_id':game_id,'player_id':player_id, 'parameter_code' :parameter_code,'whole_game' :whole_game}, con=engine)
    return df
```

```python
#half data entire game
def select_distance_half_eg():
    sql="select distance,HALF_INDICATOR,game_id\
          from distance_per_half_v\
          where game_id <>754\
          and game_id<>640"
    df = pd.read_sql(sql, con=engine)
    return df
```

```python
def Get_players (game_id,player_id):
    if game_id == '':
        game_id = '%'
    if player_id == '':
        player_id = '%'
        
    sql = "select distinct player_id from aggregated_game_data_sub_t \
           where player_id <> (select id from players_t where name like 'ball')\
           and player_id like :player_id\
           and game_id like :game_id"
    
    df = pd.read_sql(sql,params={'game_id':game_id,'player_id':player_id},con=engine)
    return df
```

```python
def select_distance_second_half():
    sql="select distance,HALF_INDICATOR,game_id\
          from distance_per_half_sub_v\
          "
    df = pd.read_sql(sql, con=engine)
    return df
```

```python
def select_distance_second_half_rnd():
    sql="select distance,HALF_INDICATOR,game_id\
          from distance_per_half_rnd_sub_v\
          "
    df = pd.read_sql(sql, con=engine)
    return df
```

```python
def select_distance_total():
    sql="select sum(distance) distance,game_id\
          from distance_per_half_sub_v\
          where game_id <>754\
          and game_id<>640\
          group by game_id"
    df = pd.read_sql(sql, con=engine)
    return df
```

```python
def moment_of_substitution():
    sql ="select half_order_column ,  count(half_order_column) as number_of_substitutes \
         from( select distinct player_id  ,game_id\
          ,id\
          , half_indicator||order_column as half_order_column\
          ,      half_indicator\
          ,      order_column\
          ,      1 substitute\
          from   AGGREGATED_GAME_DATA_SUB_T t1\
          where parameter_code = '5M'\
         and id =(select id\
                  from aggregated_game_data_sub_t t2\
                  where parameter_code = '5M'\
                 and t1.player_id =t2.player_id\
         and t1.game_id = t2.game_id\
         order by t2.game_id,t2.player_id,t2.half_indicator,t2.order_column\
         fetch first row only)\
         and order_column < 10\
         order by half_order_column)\
         group by half_order_column\
         order by half_order_column"

    df =pd.read_sql(sql, con = engine)
    return df
```
