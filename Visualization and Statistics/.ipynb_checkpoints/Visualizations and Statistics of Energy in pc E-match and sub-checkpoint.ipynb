```python
def overview_metrics_power_cat(power_cat):
     
    sql = "select sum(t1.time) "'time'"\
           ,      sum(t1.distance) "'distance'"\
           ,      sum(t1.energy) "'energy'"\
           ,      t1.power_cat\
           ,      t2.game_id\
           ,      t2.player_id\
           from power_cat_energy_in_bin_v t1\
           ,      aggregated_game_data_t t2\
           where t1.aggregated_game_data_id = t2.id\
           and t2.parameter_code = '5M'\
           and difference_start_end_prr_ind = 'T'\
           and player_id not in (select id from players_t where name in ('ball','centroid', 'Referee'))\
           and player_id not in (select player_id from keepers_games_t)\
           and order_column <=9\
           and full_game_ind = 'T'\
           and power_cat like :power_cat\
           group by t2.game_id,t1.power_cat,t2.player_id\
           order by game_id,player_id"
        
    df = pd.read_sql(sql,params={'power_cat':power_cat},con=engine)
    return df   

```

```python
#sum the total energy for all games
def total_ee_metrics():
    sql = "select    sum(t1.energy) "'energy'"\
           from power_cat_energy_in_bin_v t1\
           ,      aggregated_game_data_t t2\
           where t1.aggregated_game_data_id = t2.id\
           and t2.parameter_code = '5M'\
           and difference_start_end_prr_ind = 'T'\
           and player_id not in (select id from players_t where name in ('ball','centroid', 'Referee'))\
           and player_id not in (select player_id from keepers_games_t)\
           and full_game_ind = 'T'\
           group by\
           t2.game_id,t2.player_id\
           order by game_id,player_id"
    df = pd.read_sql(sql,con=engine)
    return df        
```

```python
#sum the totalof energy for the halfs
def total_half_ee_metrics():
    sql = "select    sum(t1.energy) energy\
          ,avg(t3.energy) avg_energy\
          , (sum(t1.energy)/avg(t3.energy)*100) percentage\
          ,      half_indicator\
           from power_cat_energy_in_bin_v t1\
           ,      aggregated_game_data_t t2\
          ,( select   sum(t1.energy) energy\
          ,      game_id\
          ,      player_id\
          from power_cat_energy_in_bin_v t1\
           ,      aggregated_game_data_t t2\
           where t1.aggregated_game_data_id = t2.id\
           and t2.parameter_code = '5M'\
           and full_game_ind='T'\
           and difference_start_end_prr_ind = 'T'\
           and player_id not in (select id from players_t where name in ('ball','centroid', 'Referee'))\
           and player_id not in (select player_id from keepers_games_t)\
           group by t2.game_id,player_id) t3\
           where t1.aggregated_game_data_id = t2.id\
           and t3.game_id = t2.game_id\
           and t3.player_id =t2.player_id\
           and t2.parameter_code = '5M'\
           and t2.full_game_ind ='T'\
           and t2.difference_start_end_prr_ind = 'T'\
           and t2.player_id not in (select id from players_t where name in ('ball','centroid', 'Referee'))\
           and t2.player_id not in (select player_id from keepers_games_t)\
           group by t2.game_id,t2.player_id, t2.half_indicator\
           order by t2.game_id,t2.player_id" 
    df = pd.read_sql(sql, con = engine)
    return df
```
